```python
import sys
import os

os.chdir('../')
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from functions import genomic_data_handling as gdh
from functions import params
from functions import data_processing as dp
from functions import lib_table_operations as lto
import seaborn.objects as so
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
rossi_lib_z = params.rossi_lib_z
chec_lib_z =  params.LIB_Z_SCORE
```

```python
z_tresh = 0
tf_tresh = 1

filtered_df = pd.DataFrame()
datasets = ['ChEC','Rossi']
for df_i, df in enumerate([chec_lib_z, rossi_lib_z]):
    z_tresh_df = df.query("Lib_occup>@z_tresh")
    for tf in z_tresh_df['TF'].unique():
        if z_tresh_df.query("TF==@tf").shape[0]>tf_tresh:
            tf_tresh_df = z_tresh_df.query("TF==@tf")
            tf_tresh_df['Dataset'] = [datasets[df_i]]*len(tf_tresh_df)
            filtered_df = pd.concat([filtered_df,tf_tresh_df])
filtered_df = filtered_df.reset_index(drop=True)
zero_binding_inds = filtered_df.query("Dataset=='Rossi' & Genomic_binding_sum == 0").index.values
filtered_df.loc[zero_binding_inds, 'Genomic_binding_log2'] = 0.0
```

## model TFs genomic comparison

```python
model_tfs = ['Tec1', 'Ste12', 'Msn2', 'Reb1', 'Abf1', 'Rap1']
rows = 1
cols= len(model_tfs)
f_size = (4*cols,3)
datasets = ['ChEC','Rossi']
y_col = 'Genomic_lib_z'
x_col = 'Lib_occup'

for dataset in datasets:
    fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=f_size)
    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4,hspace=0.4)
    fax = ax.ravel()

    df = filtered_df.query("Dataset==@dataset & TF in @model_tfs")

    y_min = df[y_col].min()
    y_max = df[y_col].max()

    x_min = 0
    x_max = 100
   
    for tf_i,tf in enumerate(model_tfs):
        tf_binding = df.query("TF==@tf")
        sns.scatterplot(data=tf_binding,x=x_col,y=y_col,hue='Promoter_name', ax=fax[tf_i])
        pearsons_corr = np.round(np.corrcoef(tf_binding[x_col].values.astype(float),tf_binding[y_col].values.astype(float))[0,1],2)
        fax[tf_i].set_title(tf+ ", Pearson's r: " + str(pearsons_corr))
        fax[tf_i].set_xlabel('')
        fax[tf_i].set_ylabel('')
        fax[tf_i].legend(loc='upper left', bbox_to_anchor=(1, 1))
```

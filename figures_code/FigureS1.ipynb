## (A) GRFs bound promoters with additional TFs

```python
tf_motifs.loc['Reb1'] = 'CGGGT'
sum_prom_zscore_thresh_main_tf = 3
sum_prom_zscore_thresh_second_tf = 3
prom_length = 1000
motif_wind = 18
min_enrichment = 2
tf_rank_thresh =  3
bound_motif_thersh = motif_wind*2/prom_length*min_enrichment 
# we expect that by chance the signal around the motif will be the size of the window/promoterlength,
#the thershold we take is "min_enrichment" times higher that by chnce.
max_mot_dist = 90
grfs = ['Reb1','Rap1','Abf1']
tf_zscore_sumprom = tf_sumprom.apply(zscore, nan_policy='omit')

fig,axes = plt.subplots(nrows=1, ncols=len(grfs), figsize=(18,6))
for i,tf in enumerate(grfs):
    tf_motif = tf_motifs.loc[tf].values[0]
    tf_zscore_sumprom.index = GP.index
    tf_bound_proms = tf_zscore_sumprom.loc[tf_zscore_sumprom[tf] > sum_prom_zscore_thresh_main_tf,tf].index
    tf_prom_count = []
    tf_on_prom = []
    tf_mot_bound_proms = []
    for curr_prom in tf_bound_proms:
        near_bound_tfs = []
        chromosome = int(GP.loc[curr_prom,:]['Chromosome'])
        chromosome_str = 'chr'+str(chromosome)
        start_pos,end_pos = int(GP.loc[curr_prom,:]['TSS_stein_Start']),int(GP.loc[curr_prom,:]['TSS_stein_End']) 
        seq = gdh.get_prom_seq(chromosome_str, start_pos,end_pos, prom_length)
        signal = gdh.get_prom_signal(tf, chromosome, start_pos,end_pos,prom_length, motif_wind)
        tf_mot_pos = np.array(gdh.find_motif_in_seq(seq, tf_motif))
        tf_mot_wind_signal = [(sum(signal[mot_pos-motif_wind:mot_pos+motif_wind]) / sum(signal)) for mot_pos in tf_mot_pos]
        tf_mot_pos_passed =tf_mot_pos[np.where(np.array(tf_mot_wind_signal)>bound_motif_thersh)[0]]
        if len(tf_mot_pos_passed)==0:
            continue
        tf_mot_bound_proms.append(curr_prom)

        relevant_tfs = tf_zscore_sumprom.loc[curr_prom, tf_zscore_sumprom.loc[curr_prom] > sum_prom_zscore_thresh_second_tf].index       
        relevant_tfs = np.setdiff1d(relevant_tfs,[tf])
        for second_tf in relevant_tfs:
            if second_tf not in tf_motifs.index:
                continue
            second_tf_motif = tf_motifs.loc[second_tf].values[0]
            second_tf_mot_pos = np.array(gdh.find_motif_in_seq(seq, second_tf_motif))
            second_tf_mot_wind_signal = [(sum(signal[mot_pos-motif_wind:mot_pos+motif_wind]) / sum(signal)) for mot_pos in second_tf_mot_pos]
            tf_pos_ranks =[]
            for pos in second_tf_mot_pos:
                tf_pos_ranks.append(gdh.tf_enrichment_rank(pos,second_tf, chromosome, start_pos,end_pos,
                                             prom_length, motif_wind, np.array(tf_motifs.index)))
           
            second_tf_mot_pos_passed_ids = (np.transpose(np.array(tf_pos_ranks)) < tf_rank_thresh) | (np.array(second_tf_mot_wind_signal)>bound_motif_thersh)
            if len(second_tf_mot_pos_passed_ids) > 0:
                second_tf_mot_pos_passed = second_tf_mot_pos[second_tf_mot_pos_passed_ids[0]]
            else:
                second_tf_mot_pos_passed = []
                                   
            if len(second_tf_mot_pos_passed)==0:
                continue

            for mot1 in tf_mot_pos_passed:
                for mot2 in second_tf_mot_pos_passed:
                    if np.abs(mot1-mot2) < max_mot_dist:
                        second_tf_signal = gdh.get_prom_signal(second_tf, chromosome, start_pos,end_pos,prom_length, motif_wind)
                        mot2_signal = (sum(second_tf_signal[mot2-motif_wind:mot2+motif_wind]) / sum(second_tf_signal))
                        if (mot2_signal > bound_motif_thersh) & (second_tf not in near_bound_tfs):
                            near_bound_tfs.append(second_tf)

        tf_prom_count.append(len(near_bound_tfs))
        tf_on_prom.append(near_bound_tfs)
    sns.histplot(tf_prom_count, ax=axes[i])
    axes[i].set_title(tf)
```

## (B) Libraries illustration presented in (C,D and S1B, S1C)

```python
cols = ['r','k','g']
grf_col_dict = {'Reb1':'k','Abf1':'r','Rap1':'g'}
rom_chr = ['I','II','III','IV','V','VI','VII','VIII','IX','X','XI','XII','XIII','XIV','XV','XVI']
lib_size = 164
prom_length = 1000
grfs = ['Reb1','Rap1','Abf1']
matching = [i for i,s in enumerate(list(libs['Relevant_factors'].values)) if any(xs in s for xs in grfs)]
rel_libs = libs.iloc[matching,:]
for lib in rel_libs.index:
    lib_info = dp.get_lib_info(lib)

    gene = lib_info['gene']
    lib_start = lib_info['start_position']
    g_start,g_end = int(GP.loc[gene.upper()]['TSS_stein_Start']),int(GP.loc[gene.upper()]['TSS_stein_End'])
    chromosome = int(GP.loc[gene.upper()]['Chromosome'])
    tfs = []
    positions = [] 
    for i,tfs_in_pos in enumerate(lib_info['mut_by_tf']):
        tfs = tfs + tfs_in_pos.split('/')
        positions = positions + [i]*len(tfs_in_pos.split('/'))
    locations = [lib_info['mut_loc'][i] for i in positions]
    grf_pos = [i for i,tf in enumerate(tfs) if tf in grfs]
    tf_types = ['STF']*len(tfs)
    for loc in grf_pos:
        tf_types[loc] = 'GRF'
    opts = np.arange(0,np.max(positions)+1)
    y_pos = np.zeros([1,len(positions)])[0]
    dat_mat = np.empty([len(tfs),lib_size])
    for i,tf in enumerate(tfs):
        if g_start < g_end:
            curr_signal = LAB_WT_NORM[tf][chromosome-1][g_start-prom_length+lib_start:
                                                          g_start-prom_length+lib_start+lib_size]
            lib_start_pos = g_start-prom_length+lib_start
            lib_end_pos =  g_start-prom_length+lib_start+lib_size
        else:
            curr_signal = LAB_WT_NORM[tf][chromosome-1][g_start+prom_length-lib_start:
                                                          g_start+prom_length-lib_start-lib_size:-1]
            lib_start_pos =  g_start+prom_length-lib_start
            lib_end_pos =  g_start+prom_length-lib_start-lib_size
        dat_mat[i,:] = curr_signal.T
    
    sum_signal = np.round(np.sum(dat_mat,axis=1))
    
    
    df = pd.concat([pd.Series(tfs),pd.Series(positions),pd.Series(locations),pd.Series(y_pos),pd.Series(sum_signal),pd.Series(tf_types)],axis=1)
    df.columns = ['tf','pos','loc','y_pos','signal','tf_type']
    df.loc[df['tf_type']=='STF','signal']=0
    grf_pos_data = []
    for grf in np.unique(df.loc[df['tf_type']=='GRF']['tf']):
        grf_pos_data.append(np.min([i for i,tf in enumerate(tfs) if tf==grf]))
    fig,ax = plt.subplots(1,1,figsize=(10,2))
    for grf_pos in grf_pos_data:
        ax.fill_between(range(dat_mat.shape[1]),dat_mat[grf_pos,:]/np.max(dat_mat[grf_pos,:]), color=grf_col_dict[tfs[grf_pos]],alpha=0.3)
    sns.scatterplot(data=df,x='loc',y='y_pos',hue='tf',size='signal',sizes=(200,400),ax=ax)
    cols = ['r','k','g']
    grf_col_dict = {'Reb1':'k','Abf1':'r','Rap1':'g'}
    sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
    ax.set_xlim(0,lib_size)
    ax.set_title(gene + ' promoter' +'\n' +'Chromosome ' + rom_chr[chromosome-1] + ' bps:' + str(lib_start_pos) + '-' + str(lib_end_pos))
```

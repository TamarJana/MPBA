## (A) Tec1 & Ste12 genomic promoter binding

```python
tfa = 'Ste12'
tfb = 'Tec1'
mota = tf_motifs.loc[tfa].values[0]
motb = tf_motifs.loc[tfb].values[0]
window = 20
pad_signal = window
bound_threshold = 2.5
prom_length = 1000
quantile = 0.75
max_mot_dist = 50
prom_ids = gdh.comb_proms(tfa,tfb,tf_zscore_sp,bound_threshold)
tfa_mot_exists = [] 
tfb_mot_exists = []  
min_dist = []
for curr_prom in prom_ids:
    signal_tfa = gdh.get_signal(curr_prom,tfa,pad_signal,prom_length)
    signal_tfb = gdh.get_signal(curr_prom,tfb,pad_signal,prom_length)
    dna_seq = gdh.get_seq(curr_prom,prom_length) 
    mota_locs = gdh.find_motif_loc(dna_seq, mota)
    motb_locs = gdh.find_motif_loc(dna_seq, motb)
    mota_bound_locs,perc_sig = gdh.get_bound_mot_locs(mota_locs,signal_tfa,pad_signal,window,quantile, prom_length)
    motb_bound_locs,_ = gdh.get_bound_mot_locs(motb_locs,signal_tfb,pad_signal,window,quantile, prom_length)
    if len(mota_bound_locs)>0:
        tfa_mot_exists.append(2)
    else:
        tfa_mot_exists.append(0)
    if len(motb_bound_locs)>0:
        tfb_mot_exists.append(1)
    else:
        tfb_mot_exists.append(0)
    if (len(mota_bound_locs)>0) & (len(motb_bound_locs)>0):
        dists = []
        for curr_a_loc in mota_bound_locs:
            for curr_b_loc in motb_bound_locs:
                dists.append(curr_a_loc-curr_b_loc)
        min_dist.append(np.min(np.abs(dists)))
    elif len(mota_bound_locs)>1:
        min_dist.append(np.min(np.abs(np.diff(np.sort(mota_bound_locs)))))
    elif len(motb_bound_locs)>1:
        min_dist.append(np.min(np.abs(np.diff(np.sort(motb_bound_locs)))))
    else:
        min_dist.append(prom_length)
min_dist = np.array(min_dist)
min_dist[min_dist>max_mot_dist] = max_mot_dist
for_size = ((max_mot_dist-min_dist)*8)+20
mot_exists = np.array(tfa_mot_exists)+np.array(tfb_mot_exists)
fig,ax = plt.subplots(1,1,figsize=(8,8))
tfa = 'Ste12'
tfb = 'Tec1'
genes = ['Pcl2','Pop3','Sim1','Cdc6','Gpa1','Kar4','Tec1']
gene_ids = []
for gene in genes:
    gene_ids.append(np.where(GP.index==gene.upper())[0][0])
plt.scatter(tf_sumprom[tfa],tf_sumprom[tfb],c='#abb0ac',s=20)
plt.scatter(tf_sumprom.loc[prom_ids,tfa],tf_sumprom.loc[prom_ids,tfb],c=mot_exists,s=for_size,cmap='tab10')
plt.colorbar()
x = tf_sumprom.loc[gene_ids,tfa].values
y = tf_sumprom.loc[gene_ids,tfb].values
for i, txt in enumerate(genes):
    plt.annotate(txt, (x[i], y[i]))
ax.set_xlabel(tfa)
ax.set_ylabel(tfb)
```

## (C) Libraries illustration, including signal and motifs

```python
lib_size = 164
prom_length = 1000
tf_of_int = ['Ste12','Tec1']
cols = ['r','k','g','c']
int_tf_col_dict = {}
for i,int_tf in enumerate(tf_of_int):
    int_tf_col_dict[int_tf] = cols[i]
matching = [i for i,s in enumerate(list(libs['Relevant_factors'].values)) if any(xs in s for xs in tf_of_int)]
rel_libs = libs.iloc[matching,:]
for lib in rel_libs.index:
    lib_info = dp.get_lib_info(lib)
    gene = lib_info['gene']
    lib_start = lib_info['start_position']
    g_start,g_end = int(GP.loc[gene.upper()]['TSS_stein_Start']),int(GP.loc[gene.upper()]['TSS_stein_End'])
    chromosome = int(GP.loc[gene.upper()]['Chromosome'])
    tfs = []
    positions = [] 
    for i,tfs_in_pos in enumerate(lib_info['mut_by_tf']):
        tfs = tfs + tfs_in_pos.split('/')
        positions = positions + [i]*len(tfs_in_pos.split('/'))
    locations = [lib_info['mut_loc'][i] for i in positions]
    tf_of_int_pos = [i for i,tf in enumerate(tfs) if tf in tf_of_int]
    tf_types = ['STF']*len(tfs)
    for loc in tf_of_int_pos:
        tf_types[loc] = 'INT'
    opts = np.arange(0,np.max(positions)+1)
    y_pos = np.zeros([1,len(positions)])[0]
    dat_mat = np.empty([len(tf_of_int),lib_size])
    for i,int_tf in enumerate(tf_of_int):
        if g_start < g_end:
            curr_signal = LAB_WT_NORM[int_tf][chromosome-1][g_start-prom_length+lib_start:
                                                          g_start-prom_length+lib_start+lib_size]
            lib_start_pos = g_start-prom_length+lib_start
            lib_end_pos =  g_start-prom_length+lib_start+lib_size
        else:
            curr_signal = LAB_WT_NORM[int_tf][chromosome-1][g_start+prom_length-lib_start:
                                                          g_start+prom_length-lib_start-lib_size:-1]
            lib_start_pos =  g_start+prom_length-lib_start
            lib_end_pos =  g_start+prom_length-lib_start-lib_size
        dat_mat[i,:] = curr_signal.T
    df = pd.concat([pd.Series(tfs),pd.Series(positions),pd.Series(locations),pd.Series(y_pos),pd.Series(tf_types)],axis=1)
    df.columns = ['tf','pos','loc','y_pos','tf_type']
    fig,ax = plt.subplots(1,1,figsize=(10,2))
    for i,int_tf in enumerate(tf_of_int):
        ax.fill_between(range(dat_mat.shape[1]),dat_mat[i,:]/np.max(dat_mat[i,:]), color=int_tf_col_dict[int_tf],alpha=0.3)
    sns.scatterplot(data=df, x='loc', y='y_pos', hue='tf', size='tf_type', size_order=['INT','STF'], linewidth=0, sizes=(100,300),ax=ax)
    sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
    ax.set_xlim(0,lib_size)
    ax.set_title(gene)
```

## Tec1-Ste12 examples - occupancy at each sequence 

```python
tp = '180'
exp_num = '13'
tfs = ['Ste12','Tec1']

libs = ['37', '96', '64']
pos_by_lib = [[4,5], [1,2], [4,5]]


for lib_i in range(len(libs)):
    fig,ax = plt.subplots(figsize=(5,5))
    lib_num = libs[lib_i]
    relevant_pos = pos_by_lib[lib_i]

    lib_info = dp.get_lib_info(lib_num)
    lib_name = lib_info['gene']
    wt_var_seq = ''.join(lib_info['wt_at_var_loc'])  # get wild-type sequence
    wt_seq_in_pos = np.array(list(wt_var_seq))[relevant_pos]  # wt seq at relevant positions
    mut_var_seq = lto.get_mut_var(lib_num)
    mut_seq_in_pos = np.array(list(mut_var_seq))[relevant_pos]

    lib_df = pd.DataFrame()
    for tf in tfs:
        res_table = dp.mean_over_bio_reps(lto.norm_res_data(tf, lib_num, exp_num, norm_to=None, occ=True) * 100)
        by_tp = res_table.filter(items=[tp])
        by_tp = by_tp.rename({tp: tf}, axis=1)
        lib_df = pd.concat([lib_df, by_tp.sort_index(ascending=False)],axis=1)
    seqs = list(res_table.index)
    all_seqs_split = np.array([list(seq) for seq in seqs])

    id_both_wt = np.sum(all_seqs_split[:, relevant_pos] == wt_seq_in_pos, axis=1) == len(relevant_pos)
    id_both_mut = np.sum(all_seqs_split[:, relevant_pos] == mut_seq_in_pos, axis=1) == len(relevant_pos)
    first_pos_wt = np.sum(all_seqs_split[:, [relevant_pos[0]]] == wt_seq_in_pos[0], axis=1) == 1
    first_pos_mut = np.sum(all_seqs_split[:, [relevant_pos[0]]] == mut_seq_in_pos[0], axis=1) == 1
    sec_pos_wt = np.sum(all_seqs_split[:, [relevant_pos[1]]] == wt_seq_in_pos[1], axis=1) == 1
    sec_pos_mut = np.sum(all_seqs_split[:, [relevant_pos[1]]] == mut_seq_in_pos[1], axis=1) == 1
    only_first_wt = first_pos_wt & sec_pos_mut
    only_sec_wt = sec_pos_wt & first_pos_mut

    min_lim = min(lib_df.min()-5)
    max_lim = max(lib_df.max()+5)

    seqs=np.array(seqs)
    lib_df['Seq_type'] = np.zeros((len(seqs),1))
    lib_df['Seq_type'].loc[seqs[id_both_wt]] = '++'
    lib_df['Seq_type'].loc[seqs[id_both_mut]] = '--'
    lib_df['Seq_type'].loc[seqs[only_first_wt]] = '+-'
    lib_df['Seq_type'].loc[seqs[only_sec_wt]] = '-+'

    ax=sns.scatterplot(data=lib_df, x=tfs[0], y=tfs[1], hue='Seq_type')

    ax.set_xlim(min_lim, max_lim)
    ax.set_ylim(min_lim, max_lim)
    ax.plot([min_lim,max_lim], [min_lim,max_lim], ls="--", c=".3")
    ax.legend(loc='upper left', bbox_to_anchor=(1, 1.1))
    ax.set_title(lib_name + ' ' + ' '.join(np.array(lib_info['mut_by_tf'])[np.array(relevant_pos)]))
    mean_df = lib_df.groupby('Seq_type')[tfs].mean()
    sns.scatterplot(data=mean_df, x=tfs[0], y=tfs[1], hue='Seq_type', s=300, legend=False)
    mean_df_frac = 1-mean_df/100
    exp = (1- mean_df_frac.loc[['+-', '-+']].product() )*100
    exp = exp.to_frame().T
    sns.scatterplot(data=exp, x=tfs[0], y=tfs[1], s=200, color='k')
    
```
